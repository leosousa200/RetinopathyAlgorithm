```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style="darkgrid")
```

# Learning Curves for Full Training Set

```python
df_full = pd.read_csv('./logger/trainlog_baseline.csv')
df_full.rename(columns = {'categorical_accuracy': 'Train',
                          'val_categorical_accuracy': 'Val',
                          'loss': 'train_loss'},
                           inplace = True )    

print(df_full.head())
print(df_full.tail())
```

```python
fig1, ax1 = plt.subplots(figsize=(9, 4))
sns.lineplot(data = df_full[['Train','Val']], ax = ax1)
ax1.set_xlabel('epoch')
ax1.set_ylabel('Accuracy')

fig2, ax2 = plt.subplots(figsize=(9, 4))
sns.lineplot(data = df_full[['train_loss','val_loss']], ax = ax2)
ax2.set_xlabel('epoch')
ax2.set_ylabel('Loss')
```

```python
df_al = pd.read_csv('./logger/trainlog_al.csv')
df_al.rename(columns = {'categorical_accuracy': 'Train',
                          'val_categorical_accuracy': 'Val',
                          'loss': 'train_loss'},
                           inplace = True )
df_al.head()
```

```python
fig3, ax3 = plt.subplots(figsize=(9, 4))
fig4, ax4 = plt.subplots(figsize=(9, 4))

sns.lineplot(data = df_al[['Train','Val']], ax = ax3)
ax3.set_xlabel('epoch')
ax3.set_ylabel('Accuracy')

sns.lineplot(data = df_al[['train_loss','val_loss']], ax = ax4)
ax4.set_xlabel('epoch')
ax4.set_ylabel('Loss')
```
