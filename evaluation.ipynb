```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style="darkgrid")
```

# Learning Curves for Full Training Set

```python
df_full = pd.read_csv('./logger/trainlog_baseline.csv')
df_full.rename(columns = {'categorical_accuracy': 'Train',
                          'val_categorical_accuracy': 'Val',
                          'loss': 'train_loss'},
                           inplace = True )    
df_full.head()
```

```python
fig1, ax1 = plt.subplots(1,2,figsize=(18, 4))

sns.lineplot(data = df_full[['Train','Val']], ax = ax1[0])
ax1[0].set_xlabel('epoch')
ax1[0].set_ylabel('Accuracy')

sns.lineplot(data = df_full[['train_loss','val_loss']], ax = ax1[1])
ax1[1].set_xlabel('epoch')
ax1[1].set_ylabel('Loss')
```

```python
df_al = pd.read_csv('./logger/trainlog_al.csv')
df_al.rename(columns = {'categorical_accuracy': 'Train',
                          'val_categorical_accuracy': 'Val',
                          'loss': 'train_loss'},
                           inplace = True )
df_al.head()
```

```python
fig2, ax2 = plt.subplots(1,2,figsize=(18, 4))

sns.lineplot(data = df_al[['Train','Val']], ax = ax2[0])
ax2[0].set_xlabel('epoch')
ax2[0].set_ylabel('Accuracy')

sns.lineplot(data = df_al[['train_loss','val_loss']], ax = ax2[1])
ax2[1].set_xlabel('epoch')
ax2[1].set_ylabel('Loss')
```

```python
df_trk = pd.read_csv('./logger/AL_tracking.csv')
df_trk.head()
```
