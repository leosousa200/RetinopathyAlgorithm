```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style="darkgrid")
```

# Learning Curves for undersample Transfer Learning Model

```python
df_under = pd.read_csv('./logger/trainlog_transferlearning_under.csv')
df_under.rename(columns = {'categorical_accuracy': 'Train under',
                          'val_categorical_accuracy': 'Accuracy with undersample',
                          'loss': 'Loss with undersample'},
                           inplace = True )    

print(df_under.head())
print(df_under.tail())
```

# Learning Curves for data augmentation Transfer Learning Model

```python
df_aug = pd.read_csv('./logger/trainlog_transferlearning_aug.csv')
df_aug.rename(columns = {'categorical_accuracy': 'Train aug',
                          'val_categorical_accuracy': 'Accuracy with augmented',
                          'loss': 'Loss with augmented'},
                           inplace = True )    
# join the two tf models
df = pd.concat([df_under, df_aug])
```

# Accuracy

```python
fig2, ax2 = plt.subplots(figsize=(9, 4))

# Plotando a linha de df_1
sns.lineplot(data=df[['Accuracy with undersample','Accuracy with augmented']], ax=ax2, linestyle='dotted')

# Plotando a linha de df_2
#ns.lineplot(data=df_aug[['Val']], ax=ax2)

# Configurando o gráfico
ax2.set_xlabel('epoch')
ax2.set_ylabel('Accuracy')
plt.legend()  # Mostra a legenda com os rótulos
plt.show()
```

# Train loss

```python
fig3, ax3 = plt.subplots(figsize=(9, 4))

# Plotando a linha de df_1
sns.lineplot(data=df[['Loss with undersample','Loss with augmented']], ax=ax3, linestyle='dotted')

# Plotando a linha de df_2
#ns.lineplot(data=df_aug[['Val']], ax=ax2)

# Configurando o gráfico
ax3.set_xlabel('epoch')
ax3.set_ylabel('Loss')
plt.legend()  # Mostra a legenda com os rótulos
plt.show()
```
